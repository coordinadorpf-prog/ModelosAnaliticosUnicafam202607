## LISTAS


```python
# Una lista vacia
mi_lista=[]
# lista con varios elementos
mi_lista=[1,2,3,4,5.0,True]
print(mi_lista[2])
```

```python
# Cambios en la lista
mi_lista[0]="dalia"
print(mi_lista)
```

```python
# Agregar elementos a una lista
mi_lista.append("zuluaga")
print(mi_lista)
```

```python
# Agregar elementos a una posicion especifica
mi_lista.insert(3,"caldas")
print(mi_lista)
```

```python
# Agregar varios elementos al final de la lista
mi_lista.extend([7,False])
print(mi_lista)
```

```python
# Quitar elementos
# Eliminar objetos
mi_lista.remove(True)
print(mi_lista)
```

```python
print("ACTUALIZACION DE DATOS")
valor_buscar=input("Por favor, ingrese el nombre para actualizar:")
valor_reemplazar=input("Ingrese valor a reemplazar:")
mi_lista[mi_lista.index(valor_buscar)]=valor_reemplazar
print("Cambio ejecutado. Imprimiendo resultado")
print(mi_lista)
```

```python
# ELIMINACION DE DATOS
valor_buscar=input("Por favor ingrese el nombre a eliminar").lower()
for item in mi_lista:
    if isinstance(item,str)and item.lower()==valor_buscar:
      mi_lista.remove(item)
      break
print(mi_lista)
```
