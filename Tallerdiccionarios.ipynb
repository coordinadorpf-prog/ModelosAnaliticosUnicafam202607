```python
# CREAR DICCIONARIO CON DATOS PROPIOS
perfil={
    "Nombre":"Dalia Zuluaga",
    "Ciudad":"Colombia",
    "Universidad":"UNICAFAM",
    "Docente":"Nicolas Mojica",
    "Profesion":"Administradora de empresas",
    "Hobbies":["leer","bailar","escuchar musica"]
}
print(perfil)
```

```python
# Menu de opciones
print("---- MENU MI PERFIL ----")
print("1. Agregar un dato nuevo")
print("2. Actualizar un dato existente")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con .pop()")
print("5. Consultar un dato con .get()")
print("6. Ver el perfil completo")
opcion=input("Elige una opcion(1-6):")
```

```python
# Opcion 1:Agregar un dato nuevo
if opcion == "1":
  clave=input("Ingresa la clave nueva:")
  valor=input("Ingresa el valor")
  perfil[clave]=valor
  print("dato agregado")
  print(perfil)
```

```python
# Opcion 2: Actualizar un dato existente
if opcion == "2":
   clave=input("Ingresa la clave para actualizar:")
   if clave in perfil:
      valor=input("Ingresa nuevo valor:")
      perfil[clave]=valor
      print("Dato actualizado.")
      print(perfil)
   else:
      print(f"La clave{clave}no existe el perfil.")
```

```python
# Opcion 3: Eliminar un dato con del
if opcion == "3":
    clave=input("Ingresar la clave a eliminar:")
    if clave in perfil:
        del perfil[clave]
        print(" Dato eliminado.")
        print(perfil)
    else:
        print(f"La clave {clave} no existe en el perfil.")
```

```python
# Opcion 4: Eliminar un dato con .pop()
if opcion == "4":
    clave=input("Ingresa la clave a eliminar:")
    if clave in perfil:
        valor_eliminado=perfil.pop(clave)
        print(f"Dato eliminado:{valor_eliminado}")
        print(perfil)
    else:
        print(f"La clave{clave}no existe en el perfil")
```
