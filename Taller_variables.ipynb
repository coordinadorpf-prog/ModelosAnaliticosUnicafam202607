**VARIABLES**

**Declaración, tipos de datos y asignación**

```python
# 1. Mis primeras variables
nombre = "Dalia"
apellido = "Zuluaga"
edad = 40
altura = 1.65
activa = True

print(nombre)
print(apellido)
print(edad)
print(altura)
print(activa)
```

```python
# 2. Predice la salida .. ¿Qué imprime esto?
x = 10
y = x
x = 20
print(x) # X->20
print(y) # Y->20
```

```python
# 3. Intercambio de variables
# Intercambia los valores de a y b sin escribir el texto directamente

a = "hola"
b = "mundo"

# Guardamos a en una variable temporal
temporal = a
a = b
b = temporal

print(a)  # mundo
print(b)  # hola
```

```python
# 4. Nombres de variables invalidos - Encuentra el error

# 1nombre = "Carlos"    # ERROR: no puede empezar con numero
nombre1 = "Carlos"      # CORRECTO

mi_apellido = "Ruiz"    # CORRECTO

# mi ciudad = "Bogota"  # ERROR: no puede tener espacio
mi_ciudad = "Bogota"    # CORRECTO

_privado = 42           # CORRECTO

# class = "matematicas" # ERROR: class es palabra reservada
clase = "matematicas"   # CORRECTO
```

```python
# 5.Conversion de tipos
edad_texto = "23"

# 1. Convertir a entero
edad_entero = int(edad_texto)
print(edad_entero)  # 23

# 2. Sumarle 10
edad_suma = edad_entero + 10

# 3. Imprimir resultado
print(edad_suma)  # 33

# 4. Convertir 3.9 a entero - elimina los decimales sin redondear
numero = int(3.9)
print(f"Al convertir 3.9 a entero el resultado es: {numero}")
```
