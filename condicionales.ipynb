```python
# verificar si una persona es mayor de edad
edad=int(input("Ingresa tu edad:"))

if edad>=18:
  print(f"Eres mayor de edad, tienes {edad} años")
else:
  print(f"Eres menor de edad, tienes {edad} años")
```

```python
# Verificar si un numero es par o impar
numero=int(input("Ingresa numero: "))
if numero % 2 == 0:
  print(f"El numero {numero} es par")
else:
  print(f"El numero {numero} es impar")
```

```python
# Clasificar la edad en categorias
edad = int(input("Ingresa tu edad: "))

if edad >= 0 and edad <= 12:
    print(f"Eres un Niño, tienes {edad} años")
elif edad >= 13 and edad <= 17:
    print(f"Eres un Adolescente, tienes {edad} años")
elif edad >= 18 and edad <= 64:
    print(f"Eres un Adulto, tienes {edad} años")
else:
    print(f"Eres un Anciano, tienes {edad} años")
```

```python
# Crea un programa que compare numeros.
numero1 = int(input("Ingresa el primer numero: "))
numero2 = int(input("Ingresa el segundo numero: "))

if numero1 > numero2:
    print(f"{numero1} es mayor que {numero2}")
elif numero1 < numero2:
    print(f"{numero2} es mayor que {numero1}")
else:
    print(f"Los dos numeros son iguales")
```

```python
# Verificar rango y divisibilidad
numero = int(input("Ingresa un numero: "))

if numero >= 1 and numero <= 100:
    if numero % 5 == 0:
        print(f"{numero} esta en el rango y es divisible por 5")
    else:
        print(f"{numero} esta en el rango pero NO es divisible por 5")
else:
    print(f"{numero} NO esta en el rango de 1 a 100")
```
