```python
# Presentacion con edad
def presentacion(nombre,edad=40):
  print(f"Hola mi nombre es {nombre} y tengo {edad} años")
presentacion("Dalia")


```

```python
# Genera una calculadora de operaciones
def operaciones(a, b):
    suma = a + b
    resta = a - b
    mult = a * b
    division = a / b
    return suma, resta, mult, division

result_suma, result_resta, result_mult, result_division = operaciones(8, 5)
print(f"Suma:{result_suma}, Resta:{result_resta}, Multiplicacion:{result_mult}, Division:{result_division}")
```

```python
# determina si una persona es mayor o menor de edad
def verificar_edad(edad):
    if edad >= 18:
        print(f"Eres mayor de edad, tienes {edad} años")
    else:
        print(f"Eres menor de edad, tienes {edad} años")

edad = int(input("Ingresa tu edad: "))
verificar_edad(edad)
```

```python
# Funcion que clasifica una nota
def clasificar_nota(nota):
    if nota >= 90 and nota <= 100:
        print("Excelente")
    elif nota >= 70 and nota <= 89:
        print("Bueno")
    elif nota >= 50 and nota <= 69:
        print("Suficiente")
    else:
        print("Insuficiente")

nota = int(input("Ingrese la nota del estudiante: "))
clasificar_nota(nota)
```

```python
# calcula Tu IMC
def imc(peso, altura):
    resultado = peso / altura ** 2
    if resultado < 18.5:
        print(f"Tu IMC es {resultado}: Bajo peso")
    elif resultado <= 24.9:
        print(f"Tu IMC es {resultado}: Normal")
    elif resultado <= 29.9:
        print(f"Tu IMC es {resultado}: Sobrepeso")
    else:
        print(f"Tu IMC es {resultado}: Obesidad")

peso = int(input("Ingresa tu peso en kg: "))
altura = int(input("Ingresa tu altura en cm: "))
imc(peso, altura / 100)
```
