```python
# 1. Predice el resultado de cada operacion sin correr el codigo

a = 17
b = 5

print(a + b)   # 22 - suma
print(a - b)   # 12 - resta
print(a * b)   # 85 - multiplicacion
print(a / b)   # 3.4 - siempre retorna float
print(a // b)  # 3 - division entera
print(a % b)   # 2 - residuo de 17 entre 5
print(a ** 2)  # 289 - a elevado al cuadrado
```

```python
# 2. Calculadora de propina
cuenta = 85000

# Calcular propina del 15%
propina = cuenta * 0.15

# Calcular total a pagar
total = cuenta + propina

# Calcular cuanto paga cada persona si son 4
por_persona = total / 4

print(f"Propina: ${propina:,.0f}")
print(f"Total a pagar: ${total:,.0f}")
print(f"Por persona: ${por_persona:,.0f}")
```

EJERCICIO 4: CONDICIONALES-if, elif, else y operadores lógicos.

```python
# 1.Que rama se ejecuta con puntaje = 72
puntaje = 72

if puntaje >= 90:
    print("Sobresaliente")
elif puntaje >= 80:
    print("Notable")
elif puntaje >= 70:
    print("Aceptable")  # Esta rama se ejecuta
elif puntaje >= 60:
    print("Suficiente")
else:
    print("Reprobado")
```

```python
# 2. Clasificador de IMC
peso = 70
altura = 1.75

imc = peso / altura ** 2

if imc < 18.5:
    print(f"IMC: {imc:.1f} - Bajo peso")
elif imc <= 24.9:
    print(f"IMC: {imc:.1f} - Normal")
elif imc <= 29.9:
    print(f"IMC: {imc:.1f} - Sobrepeso")
else:
    print(f"IMC: {imc:.1f} - Obesidad")
```

```python
# 3.Condicional con bugs - 3 errores corregidos

edad = 20

if edad == 18:       # Error 1: = asigna, debe ser == para comparar
    print("Exactamente 18")
elif edad > 18:      # Error 2: faltaban los dos puntos :
    print("Mayor de 18")
else:                # Error 3: Else con mayuscula, debe ser else
    print("Menor de 18")
```

```python
# 5.Calculadora de descuento
monto = 120000

if monto < 50000:
    descuento = 0
elif monto <= 99999:
    descuento = 0.10
elif monto <= 199999:
    descuento = 0.20
else:
    descuento = 0.30

valor_descuento = monto * descuento
total = monto - valor_descuento

print(f"Descuento: ${valor_descuento:,.0f}")
print(f"Total a pagar: ${total:,.0f}")
```

```python
# 6.FizzBuzz clasico
def fizzbuzz(n):
    if n % 3 == 0 and n % 5 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)

fizzbuzz(15)  # FizzBuzz
fizzbuzz(9)   # Fizz
fizzbuzz(10)  # Buzz
fizzbuzz(7)   # 7
```

EJERCICIO 5: CICLOS-for, while, range, break y continue.

```python
# 1. Bloque A - range(5) va de 0 a 4
for i in range(5):
    print(i)
# 0, 1, 2, 3, 4

# Bloque B - range(2, 10, 3) va de 2 a 10 de 3 en 3
for x in range(2, 10, 3):
    print(x)
# 2, 5, 8

# Bloque C - recorre cada letra de "Hola"
for letra in "Hola":
    print(letra)
# H, o, l, a
```

```python
# 2.Suma acumulada
total = 0
for i in range(1, 101):
    total += i
print(f"Suma del 1 al 100: {total}")

total_pares = 0
for i in range(2, 101, 2):
    total_pares += i
print(f"Suma de pares: {total_pares}")
```

```python
# 3. Recorrer una lista
precios = [12000, 45000, 8500, 67000, 3200]
caros = 0
for i, p in enumerate(precios):
    print(f"Item {i+1}: ${p:,.0f}")
    if p > 10000:
        caros += 1
print(f"Articulos mas de $10,000: {caros}")
```

```python
# 4. While con condicion
saldo = 1000
mes = 0
while saldo > 0:
    saldo -= 170
    mes += 1
    print(f"Mes {mes}: saldo ${saldo}")
```

```python
# 5.Break y continue
numeros = [3, 7, -2, 15, 4, -8, 9, 1]
for n in numeros:
    if n < 0:
        continue
    if n > 10:
        break
    print(n)
```

```python
# 6. Tabla de multiplicar
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i}x{j}={i*j}", end="  ")
    print()
```
